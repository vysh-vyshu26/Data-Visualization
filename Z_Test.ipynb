```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
df=pd.read_csv("/content/kidney-stone-dataset.csv")
```

```python
df.head()
```

```python
df[['ph','calc']].describe()
```

```python
ztest ,pvalue = stests.ztest(df['ph'], x2=None, value=146)
print(float(pvalue))
print(ztest)
```

```python
if pvalue<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```
